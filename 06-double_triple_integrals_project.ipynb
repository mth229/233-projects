<h2>Double integrals in polar form</h2>

<ul>
<li><p>Change the Cartesian integral into an equivalent polar integral and evaluate that:</p>
</li>
</ul>


$$
\int_{-1}^{1} \int_{-\sqrt{1-y^2}}^{0} \frac{4 \sqrt{x^2 + y^2}}{1 + x^2 + y^2} dx dy.
$$


<ul>
<li><p>Find the area enclosed by one leaf of the rose $r=12 \cos(3\theta)$.</p>
</li>
</ul>

<ul>
<li><p>Find the area of the region common to the interiors of the cardioids $r=1+\cos(\theta)$ and $r=1-\cos(\theta)$. A plot gives an indication on how one may proceed:</p>
</li>
</ul>

```julia
using SymPy, Plots

ts = range(0, stop=2pi, length=100)
t1s = range(pi + pi/4, stop=2pi, length=100)
t2s = range(pi/2, stop=pi + pi/4, length=100)
r1(t) = 1 - cos(t)
r2(t) = 1 - sin(t)
x1s = [r1(t) * cos(t) for t in t1s]
y1s = [r1(t) * sin(t) for t in t1s]
x2s = [r2(t) * cos(t) for t in t2s]
y2s = [r2(t) * sin(t) for t in t2s]

plot(x1s, y1s)
plot!(x2s, y2s)
```
